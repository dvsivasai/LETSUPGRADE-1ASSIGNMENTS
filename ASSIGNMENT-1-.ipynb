```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#QUESTION-1:- plot a line plot between a and b:
a=np.arange(40,50)
b=np.arange(50,60)
plt.plot(a,b,'bo--',linewidth=2,markersize=8)
plt.title('LINE PLOT BETWEEN a AND b')
plt.xlabel('a')
plt.ylabel('b')
plt.show()
```

```python
#QUESTION-2:-Plot a line plot showing the sales trend in company 1 and 2:
days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175]  #sales of company2
plt.title("Sales trend in company 1 and 2:(Sales Distribution)")
plt.plot(days, sales_1,'r--',linewidth=2,markersize=5,color='r',marker ='o',label='sales_1')
plt.plot(days, sales_2,'r--',linewidth=2,markersize=5,  color='b' , marker ='o',label='sales_2')
plt.xlabel("Days")
plt.ylabel("Sales")
plt.legend(loc='lower right')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#QUESTION-3.plot a Pie Chart of Indian Car companies market share % values:. max ( 6 Companies market share )
Marketshares = [24.44, 22.22, 17.56, 20.45, 26.55, 5.50]
companies = ["Honda","Maruti Suzuki", "Hyundai", "Mahindra", "Tata Moters", "Toyota"]
data = {"Indian_Car_company": companies,"market_share": Marketshares}
ss = pd.DataFrame(data)
plt.pie(Marketshares, labels = companies)
plt.show()
```

```python
#QUESTION- 4. Plot a Bar Chart with India's covid-19 Active cases ( Max 10 States )
Indian_states = ["Andhra Pradesh","Bihar","Chandigarh","Delhi","Goa","Himachal Pradesh","Jammu and Kashmir","Kerala","Ladakh","Maharashtra"]
active_cases = [27195,789,93,693,1770,1207,2709,111578,104,111622]
data = {"Covid_19_active_cases": active_cases,"Indian_states": Indian_states}
ss = pd.DataFrame(data)
plt.figure(figsize=(35, 10))
plt.title("India's covid-19 Active cases")
plt.bar(Indian_states, active_cases)
```

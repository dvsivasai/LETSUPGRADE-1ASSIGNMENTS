```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Question 1:Plot using sns.relplot, line plot, box plot on fmri data set
ss=sns.load_dataset('fmri')
ss.head()
```

```python
sns.relplot(x='timepoint',y='signal',hue='subject',style='region',size='event',data=ss)
```

```python
sns.relplot(x='timepoint',y='signal',hue='subject',style='region',size='event',kind='line',data=ss)
```

```python
sns.boxplot(x='timepoint',y='signal',hue='region',data=ss)
```
